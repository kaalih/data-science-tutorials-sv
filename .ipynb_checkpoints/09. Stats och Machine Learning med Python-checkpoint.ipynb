## Supervised Learning
De vanligaste tillämpningarna inom machine learning faller under kategorin *supervised learning* vilket innebär att man matar algoritmerna med exempeldata innehållande målvariabeln. Algoritmen studerar och lär sig vilka samband som existerar bland förklaringsvariablerna och kan på det sättet bygga en model som kan estimera målvariabeln för ännu ej observerade företeelser. ### Cross Validation
I det här exemplet ska vi titta på ett par alternativa sätt att utvärdera vår model. Den vanligaste varianten av cross validation är k-fold som delar in datan i delar för träning och validering där varje k:e värde används som validering. Indelningen av mätvärdena görs slumpmässigt i k ungefär lika stora delar. Så k-1 delar används för träning och en del för validering. Detta repeteras sedan k gånger och medelvärdet av de k prediktionsfelen används som estimat av felet.

För att visa hur detta fungerar i `Scikit-learn` behöver vi importera några bibliotek. Vi importerar metoderna `KFold` och `cross_val_score` från `sklearn.cross_validation`, algoritmen `RandomForestClassifier` från `sklearn.ensamble` och utvärderingsmetoden `f1_score` från `sklearn.metrics`.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import KFold, cross_val_score
from sklearn.metrics import f1_score
```

`KFold` skapar upp `k` antal index som vi sedan kan använda för att bygga våra train- och testdataset. Den behöver input i form av antal rader i vårt data och ett värde för `k`.

```python
n = len(X)

kf = KFold(n, n_folds=10)
```

Vi kan nu loopa igenom `kf` och titta på vad som genererats. Som vi ser nedan har metoden skapat upp 10 sektioner av datat och delat upp i train och test.

```python
k = 1

for train, test in kf:
    
    msg = 'Fold {}: Training Data: {}, Holdout Data: {}'.format(k, len(train), len(test))
    print(msg)
    k = k + 1
```

Vi kan nu iterera igenom våra folds på samma sätt som ovan och träna, scora och utvärdera en Random Forest. Vi behöver först instantiera vår algoritm och sedan skapa en lista `scores` där vi sparar resultatet för varje fold. 

Vi skriver sedan ut listan på alla `k` scores och räknar medelvärdet av dessa vilket ger oss en bra indikation på hur prediktiv vår modell är.

```python
rf = RandomForestClassifier(n_estimators=10)
scores = []

for train, test in kf:
    # Generera träningsdata
    X_train = X.ix[train]
    y_train = y.ix[train]
    
    # Generera testdata
    X_test = X.ix[test]
    y_test = y.ix[test]
    
    # Träna vår modell
    model = rf.fit(X_train, y_train)
    
    # Scoring
    y_pred = model.predict(X_test)
    
    # Utvärdera med vald metod
    score = f1_score(y_test, y_pred)
    
    # Spara resultatet för varje fold i listan scores
    scores.append(score)
    
print(scores)
print('Mean Score: {:.3f}'.format(np.mean(scores)))
```

I exemplet ovan gör vi egentligen inget annat är en rak träning och prediktion. För detta finns en specialiserad funktion som minimerar hur mycket kod vi behöver skriva. Metoden heter `cross_val_score` och används som nedan. Mycket trevligt för alla lata analytiker!

```python
rf = RandomForestClassifier(n_estimators=50)
scores = cross_val_score(rf, X, y, scoring='f1', cv=10)

print(scores)
print('Random Forest: {alg:.3f}'.format(alg=scores.mean()))
```

Med detta kan vi på ett enkelt sätt utvärdera flera algoritmer för att se hur dessa presterar jämfört med varandra för att sedan välja ut en eller två att gå vidare med. Vi importerar ett antal algoritmer och instantierar dessa med default hyperparametrar för att göra det enkelt för oss.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
```

```python
rf = RandomForestClassifier(n_estimators=100)
dt = DecisionTreeClassifier()
lr = LogisticRegression()
gb = GradientBoostingClassifier()
```

För att utvärdera dessa algoritmer kan vi skapa en dict där vi har namnet som `key` och algoritmen som `value`. Vi skapar även en tom dict `mean_scores` för att lagra våra resultat. Därefter kan vi loopa igenom vår dict och spara det genomsnitliga resultatet av vår cross-validation.

```python
algorithms = {'Random Forest' : rf,
              'Decision Tree' : dt,
              'Gradient Boosting' : gb,
              'Logistic Regression' : lr}

mean_scores = {}

for key in algorithms:
    name = key
    algorithm = algorithms[key]
    
    scores = cross_val_score(algorithm, X, y, scoring='f1', cv=10)
    mean_scores[key] = np.mean(scores)
```

```python
pd.Series(mean_scores)
```

Vi ser tydligt att vi i det här skedet kan utesluta **Logistic Regression**. Vi har dessutom en lägre score på **Decision Tree** så vi kan gå vidare med **Random Forest** och **Gradient Boosting**. Hittills har vi bara arbetat med defaultinställningar så vi kan sannolikt få en ännu bättre score om vi tweakar algoritmernas hyperparametrar.

### Optimera våra modeller med hjälp av Grid Search
Optimering av en model kan vara en väldigt manuell och tidskrävande process. I `scikit-learn` har vi tillgång till ett antal moduler för att effektivisera den här processen. Första tekniken vi ska titta på är `GridSearchCV` som vi kan importera från `sklearn.grid_search`.

```python
from sklearn.grid_search import GridSearchCV
```

Med `GridSearchCV` kan vi skapa en matris av hyperparametrar som vi kan iterera igenom för att hitta en så nära optimal konfiguration som möjligt av vår algoritm. 

**VIKTIGT: ** Ta i beaktande att metoden kommer att innebär att algoritmen kommer att köras med cross-validation så varje unik kombination av parametrar kommer att köras `k` antal gånger. Ett bra tips är därför att inte köra detta på full datavolym. Exempelvis kan vi sampla data för detta ändamål.

I det enkla exemplet nedan skapar vi en dict där vi lägger in de parametrar vi vill söka igenom. I det här fallet enbart `n_estimators` som är parametern som styr hur många träd vår Random Forest ska växa.

```python
grid = {'n_estimators' : [1, 5, 25, 50, 100]}

gs = GridSearchCV(rf, grid, 'f1', cv=10)
```

Vi kör sedan vår sökning med metoden `.fit()`.

```python
search = gs.fit(X, y)
```

Vi kan nu titta på våra resultat och ta ut bäst parameter och bäst score. Vi ser att vi får ett tydligt lyft ju fler träd vi adderar även om ökningen ser ut att bli föremål för diminishing returns för högre `n`.

```python
search.grid_scores_
```

```python
print(search.best_params_)
print(search.best_score_)
```

### Receiver Operating Characteristics (ROC) och Area Under Curve (AUC)
ROC-kurvor är ett mycket bra sätt att visualisera resultatet av en binär klassificering som vi har att göra med här. Kurvan visar på ett tydligt sätt hur stort lyft en modell ger jämfört med slumpen. Mätetalet *AUC* är ett mätetal som helt enkelt räknar ut området under kurvan vilket då innebär att högt värde är bra.

För detta importerar vi funktionerna `roc_curve` och `roc_auc_score` från `sklearn.metrics`. 

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

Vi börjar med att göra en train/test split av vårt data precis som vi gjort tidigare. Vi håller i det här fallet 30% av vårt data för utvärdering. Därefter instantierar vi vår modell med de parametrar vi hittade i vår grid search och tränar den på träningsdatat.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
rf = RandomForestClassifier(n_estimators=50, max_features=5, min_samples_leaf=1)
```

```python
model = rf.fit(X_train, y_train)
```

En av fördelarna med Random Forest är att den ger en väldigt god uppskattning av styrkan hos de ingående variablerna. Vi kan titta på detta genom metoden `.feature_importances_`. Vi kan lägga in dessa i en data fram som vi sedan kan visualisera för en bättre översikt. 

```python
model.feature_importances_
```

```python
feature_importance = pd.DataFrame(model.feature_importances_, index=X.columns, columns=['importance'])
feature_importance.sort_values('importance', ascending=False, inplace=True)
```

```python
feature_importance
```

```python
p = sns.barplot(x='importance', y=feature_importance.index, data=feature_importance, color='#5B8894')
```

Vi är nu redo att köra modellen på vårt träningsdata. I det här fallet kan vi använda både `.predict()` som försöker sätta varje observation till rätt etikett och `.predict_proba()` för att uppskatta sannolikheten för varje etikett. 

Eftersom vi vill arbeta sannolikheten till churn kan vi spara detta i variabeln `y_churned`.

```python
y_prob = model.predict_proba(X_test)
y_pred = model.predict(X_test)
```

```python
y_churned = y_prob[:, 1]
```

Vi kan nu plotta vår ROC-kurva med hjälp av funktionen `roc_curve()`. Funktionen levererar ut en kurva som vi sparar i variablerna `fpr` och `tpr`. Vi får även en array med de thresholds som motsvarar respektive punkt.

```python
fpr, tpr, thresholds = roc_curve(y_test, y_churned)
```

```python
plt.plot([0,1], [0,1], 'k')
p = plt.plot(fpr, tpr, '#5B8894')
```

Vi kan nu räkna ut **AUC** och även bygga en classification report. Vi får ett mycket bra mått på AUC men om vi tittar i vår classification report så får vi ett sämre värde på `recall` vilket tyder på att vi fångar en mindre del av total churn. Detta beror på att metoden predict kräver en sannolikhet över 50% för att klassa en observation som positiv.

```python
auc = roc_auc_score(y_test, y_churned)

msg = 'Area under curve: {auc:.2f}'.format(auc=auc)
print(msg)
```

```python
print(classification_report(y_test, y_pred))
```

I det här fallet kan vi använda vår sannolikhetsvariabel för att skapa en prediktion med annan tolerans för klassificeringen. Om vi sätter gränsvärdet till 0.4 istället och kör en ny classification report ser vi att vi ökar vår recall på bekostnad av precision.

```python
threshold = 0.36
y_pred2 = np.where(y_churned > threshold, 1, 0)

print(classification_report(y_test, y_pred2))
```

```python
cm = confusion_matrix(y_test, y_pred2, labels=[0,1])
```

```python
cm = pd.DataFrame(cm, index=['no churn', 'churned'], columns=['no churn', 'churned'])
cm
```

```python
p = sns.heatmap(cm, annot=True, fmt='3d', cbar=False).set(xlabel='predicted', ylabel='true')
```

En deciluppställning kan vi göra på det här sättet.

```python
df = pd.DataFrame(list(zip(y_test, y_churned)), columns=['actual', 'prob']).sort_values('prob', ascending=True).reset_index()
df.tail()
```

```python
dummies = pd.get_dummies(df['actual'], prefix='churn')
df = pd.concat([df, dummies], axis=1)
df.head(2)
```

```python
deciles = pd.qcut(df.index, 20, labels=False) + 1
df['decile'] = deciles
```

```python
d = {'actual' : np.mean, 'churn_0' : sum, 'churn_1' : sum}

result = df.groupby('decile').agg(d)

result.columns = ['no churn', 'churned', 'ratio']

print(result.sort_index(ascending=False))
print('\ndataset mean: {m:.2f}'.format(m=y.mean()))
```
